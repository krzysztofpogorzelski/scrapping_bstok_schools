```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_hs= pd.read_csv(r"C:\Users\biuro\pad1903\własne działania\high_schools.csv", usecols=['ranking_bstok', 'ranking_pl', 'points', 'high_school', 'link'])
df_hs
```

```python
df_ts= pd.read_csv(r"C:\Users\biuro\pad1903\własne działania\technical_schools.csv", usecols=['ranking_bstok', 'ranking_pl', 'points', 'technical_school', 'link'])
df_ts
```

```python
df_ps= pd.read_csv(r"C:\Users\biuro\pad1903\własne działania\primary_schools.csv", usecols=['ranking_bstok', 'ranking_pl', 'points', 'primary_school', 'link'])
df_ps
```

```python
plt.figure(figsize=(14,7))
plt.bar(df_hs.ranking_bstok[0:12], df_hs.points[0:12], width=0.4, color="green", label=df_hs.high_school[0:12], align='center')
plt.bar(df_ts.ranking_bstok[0:12], df_ts.points[0:12], width=0.4, color="red", label=df_ts.technical_school[0:12], align='edge')
plt.bar(df_ps.ranking_bstok[0:12], df_ps.points[0:12], width=0.2, color="blue", label=df_ps.primary_school[0:12], align='edge')

plt.grid(axis="y", color="grey")
plt.xlabel('school no in ranking', fontsize=18)
plt.ylabel('points',fontsize=18)
plt.title('high, technical and primary school points', fontsize=18 )
plt.legend(loc="best", bbox_to_anchor=(1.1, 0., 0, 0.), ncol=3)
plt.savefig("points_hs_ts_ps.png", dpi=300)
plt.show()
```
