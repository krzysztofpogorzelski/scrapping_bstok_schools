## result

```python
result=[]
for row in rows:
    ranking_bstok = row.find("td").text.strip()
    ranking_pl = row.find_all("td")[1].text.strip()
    points = row.find_all("td")[3].text.strip().replace(" pkt","")
    high_school = row.find("a").text.strip()
    link= "https://waszaedukacja.pl/"+str(row.find("a").get("href"))
        
    result.append({"ranking_bstok":ranking_bstok, "ranking_pl":ranking_pl, "points":points, "high_school":high_school, "link":link})
```

```python
df_result_hs=pd.DataFrame(result)
df_result_hs
```

```python
df_result_hs.dtypes
```

```python
df_result_hs.ranking_bstok= pd.to_numeric(df_result_hs.ranking_bstok)
df_result_hs.ranking_pl= pd.to_numeric(df_result_hs.ranking_pl)
df_result_hs.points= pd.to_numeric(df_result_hs.points)
df_result_hs.dtypes
```
